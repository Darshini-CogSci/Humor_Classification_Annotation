```python
import pandas as pd
import random

# -----------------------------------
# output file
# -----------------------------------
output_file = "./final_output.xlsx" # path to file for main study

# -----------------------------------
# read files
# -----------------------------------
df1 = pd.read_csv(
    "./path/to/your/xlsx/train/dataset"
) # path to train_reddit dataset from Weller & Seppi (2019)

df2 = pd.read_excel(
    "./path/to/your/xlsx/pilot/dataset" 
) # path to pilot dataset to remove repeated jokes

# -----------------------------------
# original counts in Column B
# -----------------------------------
df1.columns = range(len(df1.columns))

original_0 = len(df1[df1[1] == 0])
original_1 = len(df1[df1[1] == 1])

print("Original rows with 0 in Column B:", original_0)
print("Original rows with 1 in Column B:", original_1)
print("Original total rows:", len(df1))

# -----------------------------------
# remove first 100 rows
# -----------------------------------
if len(df1) > 100:
    df1 = df1.iloc[100:].reset_index(drop=True)

# reset columns again
df1.columns = range(len(df1.columns))
df2.columns = range(len(df2.columns))

before_processing = len(df1)

# -----------------------------------
# remove [removed] and [deleted]
# -----------------------------------
df1 = df1[
    ~df1[3].astype(str).str.contains(r'\[removed\]|\[deleted\]', case=False, na=False)
]

# -----------------------------------
# remove by word count
# -----------------------------------
def count_words(x):
    return len(str(x).split())

df1 = df1[df1[3].apply(count_words) >= 9]
df1 = df1[df1[3].apply(count_words) <= 19]

# -----------------------------------
# remove duplicate rows
# -----------------------------------
df1 = df1.drop_duplicates(subset=[3])

# -----------------------------------
# remove rows matching second file
# -----------------------------------
check_list = df2[0].astype(str).str.strip().tolist()

df1 = df1[~df1[3].astype(str).str.strip().isin(check_list)]

# -----------------------------------
# counts after processing
# -----------------------------------
remaining_total = len(df1)

remaining_0 = len(df1[df1[1] == 0])
remaining_1 = len(df1[df1[1] == 1])

removed_total = before_processing - remaining_total
removed_0 = original_0 - remaining_0
removed_1 = original_1 - remaining_1

print()
print("After processing:")
print("Remaining rows with 0:", remaining_0)
print("Remaining rows with 1:", remaining_1)
print("Remaining total rows:", remaining_total)

print()
print("Removed during processing:")
print("Removed rows with 0:", removed_0)
print("Removed rows with 1:", removed_1)
print("Removed total rows:", removed_total)

# -----------------------------------
# sample 750 each
# -----------------------------------
group0 = df1[df1[1] == 0]
group1 = df1[df1[1] == 1]

sample0 = group0.sample(n=min(750, len(group0)), random_state=1)
sample1 = group1.sample(n=min(750, len(group1)), random_state=1)

texts0 = sample0[3].tolist()
texts1 = sample1[3].tolist()

all_texts = texts0 + texts1
random.shuffle(all_texts)

# -----------------------------------
# save output
# -----------------------------------
out = pd.DataFrame({"Text": all_texts})
out.to_excel(output_file, index=False)

print()
print("Finished. Output saved as", output_file)
```
